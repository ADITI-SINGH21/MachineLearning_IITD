# Q2

## Normalising Dataset

```python
x_1 = pd.concat([data1['Variable 1'],data1['Variable 2']],axis =1)
y_1 = data1['Class']
x1 = np.array(x_1)
y1 = np.array(y_1)
x1_train,x1_test,y1_train,y1_test =  train_test_split(x_1, y_1, test_size=0.3, random_state=42)
```

```python
autoscaler = StandardScaler()
scaler = autoscaler.fit(x_1)
x1_norm = scaler.transform(x_1)
x1_train_norm = scaler.transform(x1_train)
x1_test_norm = scaler.transform(x1_test)
y1_train = np.array(y1_train)
y1_test = np.array(y1_test)
```

## SVM Code

```python
#SVM 
def SVM(x,y,alpha1,lambda1,iteration1):
    weights = np.ones(x.shape[1])
    b = 0
    for i in range(iteration1):
        for index, arr in enumerate(x):
            check = y[index] * (np.dot(arr, weights) - b)
            if check>=1:
                weights -= alpha1 * (2 * lambda1 * weights)
            else:
                weights -= alpha1 * (2 * lambda1 * weights - np.dot(arr, y[index]))
                b -= alpha1 * y[index] 
    return weights,b
def predict(x,weights,b):
    y = np.dot(x,weights)-b
    y_pred = np.sign(y)
    return y_pred
def totalCorrect(y1,y2):
    c=0
    for i in range(len(y1)):
        if y1[i]==y2[i]:
            c+=1
    return c
```

## Checking accuracy of SVM

```python
alpha1=0.001
lambda1=0.001
iteration1=10000
weights1,b1 = SVM(x1_train_norm,y1_train,alpha1,lambda1,iteration1)
y1_pred = predict(x1_test_norm,weights1,b1)
print('the weights are',weights1)
print('the bias is',-b1)
print("number of correctly predicted labels out of 15 for test set is",totalCorrect(y1_test,y1_pred))
```

## Fitting Data 1 and plotting Decision Boundary

```python
weights1_final,b1_final = SVM(x1,y1,0.001,0.001,10000)
print('the weights are',weights1_final)
print('the bias is',-b1_final)
```

```python
fig, ax = plt.subplots()

# get the separating hyperplane
w = weights1_final
a = -w[0] / w[1]
xx = np.linspace(-5, 5)
yy = a * xx - (-b1_final) / w[1]

# create a mesh to plot in
x_min, x_max = x1[:, 0].min() - 1, x1[:, 0].max() + 1
y_min, y_max = x1[:, 1].min() - 1, x1[:, 1].max() + 1
xx2, yy2 = np.meshgrid(np.arange(x_min, x_max, .2),
                     np.arange(y_min, y_max, .2))
ax.scatter(x1[:, 0], x1[:, 1], c=y1, cmap=plt.cm.coolwarm, s=25)
ax.plot(xx,yy)

ax.axis([x_min, x_max,y_min, y_max])
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.title('Linear Decision Boundary of SVM for Data1')
plt.show()
```

```python
x_2 = pd.concat([data2['Variable 1'],data2['Variable 2']],axis =1)
y_2 = data2['Class']
x2 = np.array(x_2)
y2= np.array(y_2)
x2_train,x2_test,y2_train,y2_test =  train_test_split(x_2, y_2, test_size=0.3, random_state=42)
x2_train = np.array(x2_train)
x2_test = np.array(x2_test)
y2_train = np.array(y2_train)
y2_test = np.array(y2_test)
```

```python
from cvxopt import matrix as cvxopt_matrix
from cvxopt import solvers as cvxopt_solvers
```

```python
def OptimizedSVM(x,y,C):
    m,n = x.shape
    y = y.reshape(-1,1) * 1.
    x_dash = y * x
    H = np.dot(x_dash , x_dash.T) * 1.

    #Converting into cvxopt format - as previously
    P = cvxopt_matrix(H)
    q = cvxopt_matrix(-np.ones((m, 1)))
    G = cvxopt_matrix(np.vstack((np.eye(m)*-1,np.eye(m))))
    h = cvxopt_matrix(np.hstack((np.zeros(m), np.ones(m) * C)))
    A = cvxopt_matrix(y.reshape(1, -1))
    b = cvxopt_matrix(np.zeros(1))

    #Run solver
    sol = cvxopt_solvers.qp(P, q, G, h, A, b)
    alphas = np.array(sol['x'])
    weights = ((y * alphas).T @ x).reshape(-1,1)
    S = (alphas > 1e-4).flatten()
    b = y[S] - np.dot(x[S], weights)
    return weights.flatten(),b[0]
def predictOptimum(x,weights,b):
    y = np.dot(x,weights)+b
    y_pred = np.sign(y)
    return y_pred
```

```python
Cvalues=[0.05,0.1,1,2,5,10,20,100]
accuracy=[]
for i in range(len(Cvalues)):
    w2,b = OptimizedSVM(x2_train,y2_train,Cvalues[i])
    y2_pred = predictOptimum(x2_test,w2,b)
    accuracy.append(totalCorrect(y2_pred,y2_test)/len(y2_test))
```

## Fitting Data2 and finding Decision Boundary

```python
def printPlot(w,b,x,y):
    fig, ax = plt.subplots()
    a = -w[0] / w[1]
    xx = np.linspace(-5, 5)
    yy = a * xx - (b) / w[1]

    # create a mesh to plot in
    x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1
    y_min, y_max = x[:, 1].min() - 1, x[:, 1].max() + 1
    xx2, yy2 = np.meshgrid(np.arange(x_min, x_max, .2),
                         np.arange(y_min, y_max, .2))
    ax.scatter(x[:, 0], x[:, 1], c=y, cmap=plt.cm.coolwarm, s=25)
    ax.plot(xx,yy)

    ax.axis([x_min, x_max,y_min, y_max])
    plt.xlabel('Variable 1')
    plt.ylabel('Variable 2')
    plt.title('Linear Decision Boundary of SVM for Data1')
    plt.show()
```

```python
w2_0 = []
w2_1 = []
b2=[]
for i in range(len(Cvalues)):
    w2,b = OptimizedSVM(x2,y2,Cvalues[i])
    w2_0.append(w2[0])
    w2_1.append(w2[1])
    b2.append(b[0])
    print('The decision Boundary for C =',Cvalues[i])
    printPlot(w2,b,x2,y2)
```

```python
df=pd.DataFrame(
    {'C': Cvalues,
     'b': b2,
     'weight[0]': w2_0,
     'weight[1]': w2_1,
     'accuracy of Test Set': accuracy
    })
```
