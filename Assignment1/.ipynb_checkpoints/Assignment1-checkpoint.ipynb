## Q2(a)

### Batch LMS

```python
def batchLMS(x,y,alpha):
    final = [0,0,0,0]
    theta_0 = 1.0
    theta_1 = 1.0
    temp = 0.1
    J = []
    cost = 0.0
    n = float(len(y))
    iter = 1
    while(abs(temp-cost)>10**(-7)):
        temp = cost
        y_calc = theta_1*x+theta_0
        cost = sum((y-theta_0-theta_1*x)**2)
        cost*=0.5
        J.append(cost)
        D_1 = (-2/n) * sum(x * (y - y_calc))
        D_0 = (-2/n) * sum((y - y_calc))
        theta_0=theta_0-alpha*D_0
        theta_1=theta_1-alpha*D_1
        iter+=1
    final[0]=theta_0
    final[1]=theta_1
    final[2]=J
    final[3]=iter
    return final
```

```python
arr=batchLMS(x,y,0.01)
t0 = arr[0]
t1 = arr[1]
print(t0,t1)
```

```python
y_batch = t1*x+t0
plt.scatter(x,y)
plt.plot(x,y_batch,'r')
plt.title('Batch LMS')
```

```python
J = arr[2]
num_iter = arr[3]
iterations=np.arange(1,num_iter,1)
plt.plot(iterations,J)
plt.title('Cost vs Iterations for Batch LMS')
```

### Stochastic LMS

```python
def stochasticLMS(x,y,alpha):
    final = [0,0,0,0]
    theta_0 = 1.0
    theta_1 = 1.0
    temp = 0.1
    J = []
    cost = 0.0
    n = float(len(y))
    iter = 1
    while(abs(temp-cost)>10**(-7)):
        temp = cost
        y_calc = theta_1*x+theta_0
        for i in range(int(n)):
            D_0 = (-2/n)*(y[i]-y_calc[i])
            D_1 = (-2/n)*(x[i]*(y[i]-y_calc[i]))
            theta_0-=alpha*D_0
            theta_1-=alpha*D_1  
        cost = sum((y-theta_0-theta_1*x)**2)
        cost*=0.5
        J.append(cost)
        iter+=1
    final[0]=theta_0
    final[1]=theta_1
    final[2]=J
    final[3]=iter
    return final
```

```python
alpha2 = 0.001
arr2 = stochasticLMS(x,y,alpha2)
t0_s=arr2[0]
t1_s= arr2[1]
print(t0_s,t1_s)
```

```python
y_stochastic = t1_s*x+t0_s
plt.scatter(x,y)
plt.plot(x,y_stochastic,'r')
plt.title('Stochastic LMS')
```

```python
J_s = arr2[2]
num_iter_s = arr2[3]
iterations2=np.arange(1,num_iter_s,1)
plt.plot(iterations2,J_s)
plt.title('Cost vs Iterations for Stochastic LMS')
```

### Least Square Closed Form

```python
def LeastSquare(x,y):
    X = np.vstack([np.ones(len(y)),x])
    X_T = X.transpose()
    prod1 = np.dot(X,X_T)
    prod2 = np.dot(X,y)
    inv = np.linalg.inv(prod1)
    theta = np.dot(inv,prod2)
    return theta
```

```python
theta = LeastSquare(x,y)
print(theta)
```

```python
y_leastsquare = theta[1]*x+theta[0]
plt.scatter(x,y)
plt.plot(x,y_leastsquare,'r')
plt.title('Least Square Closed form')
```

## Q3(a)

### Logistic Regression

```python
def sigmoid(z):
    Gz = 1.0 / (1.0 + np.exp(-z))
    return Gz
def costFunction(x,y,theta):
    h = sigmoid(x.dot(theta))
    J = (-y).T.dot(np.log(h))-((1-y).T.dot(np.log(1-h)))
    J/=len(y)
    return J
def gradFunction(x,y,theta):
    h = sigmoid(x.dot(theta))
    grad = (h-y.reshape((len(y),1))).T.dot(x)
    grad/=len(y)
    return grad
def LogisticRegression(x,y,alpha,iterations):
    theta = np.zeros((3,1))
    cost = 0.0
    temp = 1.0
    J =[]
    X = np.hstack((np.ones((x.shape[0],1)),x))
    for i in range(iterations):
        temp = cost
        cost = costFunction(X,y,theta)
        grad = gradFunction(X,y,theta)
        theta-=alpha*grad.T 
        J.append(cost)
    return theta,cost,J
```

```python
def hypothesis(x,theta):
    X = np.hstack((np.ones((x.shape[0],1)),x))
    hfinal = sigmoid(X.dot(theta))
    return hfinal
def predictY(y):
    b = (y>=0.5)
    ypred=b.astype(int)
    return ypred
def totalCorrect(y1,y2):
    c=0
    for i in range(len(y1)):
        if y1[i]==y2[i]:
            c+=1
    return c
```

```python
#checking accuracy for train test values
theta1,cost1,J1 = LogisticRegression(x_l_train_norm,y_l_train,0.0075,7000)
print(theta1)
h1 = hypothesis(x_l_test_norm,theta1)
yl1 = predictY(h1)
print(yl1)
print(totalCorrect(yl1,y_l_test))
print(cost1)
```

```python
alpha_l = 0.0075
iterations_l = 7000
theta_final,cost_final,J_final = LogisticRegression(xl_norm,y_l,alpha_l,iterations_l)
H = hypothesis(test_norm,theta_final)
y_l_final = predictY(H)
```
