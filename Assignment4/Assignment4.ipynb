## Q2(b)

```python
def kmeansCluster(x,k,iterations):
    index = np.random.choice(x.shape[0], k, replace=False)
    mu = x[index, :]
    dist = cdist(x, mu ,'euclidean') 
    Rnk = np.array([np.argmin(i) for i in dist])
    for j in range(iterations): 
        mu = []
        for idx in range(k):
            temp = x[Rnk==idx].mean(axis=0) 
            mu.append(temp)
 
        mu = np.vstack(mu)
         
        dist = cdist(x, mu ,'euclidean')
        Rnk = np.array([np.argmin(i) for i in dist])
         
    return Rnk,mu
```

```python
def GMM(X,k,iterations):
    reg_cov = 1e-6*np.identity(len(X[0]))
    x,y = np.meshgrid(np.sort(X[:,0]),np.sort(X[:,1]))
    XY = np.array([x.flatten(),y.flatten()]).T
    mu = np.random.randint(min(X[:,0]),max(X[:,0]),size=(k,len(X[0])))
    cov = np.zeros((k,len(X[0]),len(X[0])))
    for dim in range(len(cov)):
        np.fill_diagonal(cov[dim],5)
    pi = np.ones(k)/k
    log_likelihoods = []
    fig = plt.figure(figsize=(4,4))
    ax0 = fig.add_subplot(111)
    ax0.scatter(X[:,0],X[:,1])
    ax0.set_title('Initial Stage')
#     fig.savefig('GMM_Initial.png')
    for m,c in zip(mu,cov):
        c += reg_cov
        multi_normal = multivariate_normal(mean=m,cov=c)
        ax0.contour(np.sort(X[:,0]),np.sort(X[:,1]),multi_normal.pdf(XY).reshape(len(X),len(X)),colors='green',alpha=0.3)
        ax0.scatter(m[0],m[1],c='grey',zorder=10,s=100)
    for i in range(iterations):
        #E_step
        r_ic = np.zeros((len(X),len(cov)))
        for m,co,p,r in zip(mu,cov,pi,range(len(r_ic[0]))):
            co+=reg_cov
            mn = multivariate_normal(mean=m,cov=co)
            r_ic[:,r] = p*mn.pdf(X)/np.sum([pi_c*multivariate_normal(mean=mu_c,cov=cov_c).pdf(X) for pi_c,mu_c,cov_c in zip(pi,mu,cov+reg_cov)],axis=0)
        mu = []
        cov = []
        pi = []
        log_likelihood = []
        #M_Step
        for c in range(len(r_ic[0])):
            m_c = np.sum(r_ic[:,c],axis=0)
            mu_c = (1/m_c)*np.sum(X*r_ic[:,c].reshape(len(X),1),axis=0)
            mu.append(mu_c)
            cov.append(((1/m_c)*np.dot((np.array(r_ic[:,c]).reshape(len(X),1)*(X-mu_c)).T,(X-mu_c)))+reg_cov)
            pi.append(m_c/np.sum(r_ic))
        log_likelihoods.append(np.log(np.sum([k*multivariate_normal(mu[i],cov[j]).pdf(X) for k,i,j in zip(pi,range(len(mu)),range(len(cov)))])))
    fig3 = plt.figure(figsize=(4,4))
    ax2 = fig3.add_subplot(111)
    ax2.scatter(X[:,0],X[:,1])
    for m,c in zip(mu,cov):
        multi_normal = multivariate_normal(mean=m,cov=c)
        ax2.contour(np.sort(X[:,0]),np.sort(X[:,1]),multi_normal.pdf(XY).reshape(len(X),len(X)),colors='black',alpha=0.3)
        ax2.scatter(m[0],m[1],c='grey',zorder=10,s=100)
        ax2.set_title('Final state')
        ax2.set_xlabel('Result 1')
        ax2.set_ylabel('Result 2')
#     fig3.savefig('GMM_final.png')
    return mu,cov
    
```
